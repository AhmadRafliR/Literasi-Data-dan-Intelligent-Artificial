```python
import pandas as pd

print("=== BAB 6: STATISTIK DESKRIPTIF & INTERPRETASI OTOMATIS ===\n")

# 1. Load Data
df_train = pd.read_csv('../data/processed/volve_train_clean.csv')

feature_cols = [
    'BORE_OIL_VOL', 'BORE_GAS_VOL', 'BORE_WAT_VOL', 
    'AVG_DOWNHOLE_PRESSURE', 'AVG_WELLHEAD_PRESSURE', 'AVG_TEMPERATURE'
]
target_col = 'AVG_CHOKE_SIZE_P'

# 2. Menghasilkan Tabel Statistik Deskriptif
deskriptif = df_train[feature_cols + [target_col]].describe()
print("1. Tabel Statistik Deskriptif:")
display(deskriptif)

# 3. GENERATE INTERPRETASI OTOMATIS BERDASARKAN ANGKA DI TABEL
print("\n2. Interpretasi Keteknikan (Otomatis Dihasilkan oleh Python):")
print("=" * 70)

# Mengambil angka spesifik dari tabel 'deskriptif'
avg_oil = deskriptif.loc['mean', 'BORE_OIL_VOL']
max_oil = deskriptif.loc['max', 'BORE_OIL_VOL']
avg_wat = deskriptif.loc['mean', 'BORE_WAT_VOL']

avg_bhp = deskriptif.loc['mean', 'AVG_DOWNHOLE_PRESSURE']
avg_whp = deskriptif.loc['mean', 'AVG_WELLHEAD_PRESSURE']

avg_choke = deskriptif.loc['mean', 'AVG_CHOKE_SIZE_P']
min_choke = deskriptif.loc['min', 'AVG_CHOKE_SIZE_P']
max_choke = deskriptif.loc['max', 'AVG_CHOKE_SIZE_P']

# Mencetak kalimat yang siap masuk ke laporan Word
print(f"Berdasarkan tabel statistik deskriptif di atas, didapatkan insight operasional sebagai berikut:")

print(f"\na. Kinerja Produksi:")
print(f"   Rata-rata produksi minyak harian sumur adalah {avg_oil:.2f} bbl/d, dengan capaian produksi ")
print(f"   maksimal (puncak) menyentuh angka {max_oil:.2f} bbl/d. Rata-rata produksi air ")
print(f"   yang terangkat bersama minyak (water cut) berada di kisaran {avg_wat:.2f} bbl/d.")

print(f"\nb. Kondisi Tekanan (Pressure):")
print(f"   Rata-rata tekanan pada dasar sumur (Downhole Pressure) tercatat sebesar {avg_bhp:.2f} psi.")
print(f"   Sesuai dengan prinsip dinamika aliran fluida dari reservoir menuju permukaan, ")
print(f"   tekanan ini mengalami penurunan (pressure drop) menjadi rata-rata {avg_whp:.2f} psi ")
print(f"   sesampainya di kepala sumur (Wellhead).")

print(f"\nc. Operasi Choke (Parameter Kontrol):")
print(f"   Sepanjang histori data yang dianalisis, operator lapangan mengatur bukaan choke ")
print(f"   pada rentang {min_choke:.2f}% hingga {max_choke:.2f}%, dengan rata-rata konfigurasi ")
print(f"   bukaan harian berada di level {avg_choke:.2f}%.")
print("=" * 70)
```
