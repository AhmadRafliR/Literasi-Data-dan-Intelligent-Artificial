```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("=== BAB 8: ANALISIS KORELASI ===")

df_train = pd.read_csv('../data/processed/volve_train_clean.csv')

feature_cols = [
    'BORE_OIL_VOL', 'BORE_GAS_VOL', 'BORE_WAT_VOL', 
    'AVG_DOWNHOLE_PRESSURE', 'AVG_WELLHEAD_PRESSURE', 'AVG_TEMPERATURE'
]
target_col = 'AVG_CHOKE_SIZE_P'

plt.figure(figsize=(10, 8))
correlation_matrix = df_train[feature_cols + [target_col]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi Pearson Fitur vs Ukuran Choke', fontweight='bold')
plt.show()
```
