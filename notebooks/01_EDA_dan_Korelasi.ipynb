```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Pengaturan tema visualisasi
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# 1. Memuat Data Train Bersih
df_train = pd.read_csv('../data/processed/volve_train_clean.csv', sep=';', decimal=',')

# KODE BARU YANG BENAR:
if 'DATEPRD' in df_train.columns:
    # Kita beritahu Python format persisnya: Tahun-Bulan-Tanggal Jam.Menit.Detik
    df_train['DATEPRD'] = pd.to_datetime(df_train['DATEPRD'], format='%Y-%m-%d %H.%M.%S')

target_col = 'AVG_CHOKE_SIZE_P'
feature_cols = [
    'BORE_OIL_VOL', 'BORE_GAS_VOL', 'BORE_WAT_VOL', 
    'AVG_DOWNHOLE_PRESSURE', 'AVG_WELLHEAD_PRESSURE', 'AVG_TEMPERATURE'
]

# 2. Statistik Deskriptif Lengkap
print("Statistik Deskriptif Variabel Produksi (Min, Max, Mean, Std, Persentil):")
display(df_train[feature_cols + [target_col]].describe())

# 3. Visualisasi Data (8 Grafik, 4 Kategori)

# Grafik 1: Heatmap Matriks Korelasi (Relationship)
plt.figure(figsize=(10, 8))
correlation_matrix = df_train[feature_cols + [target_col]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Grafik 1: Heatmap Korelasi Pearson Fitur vs Ukuran Choke')
plt.show()

# Grafik 2 & 3: Scatter Plot (Relationship)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.scatterplot(x='AVG_WELLHEAD_PRESSURE', y='AVG_CHOKE_SIZE_P', data=df_train, alpha=0.5, color='coral', ax=axes[0])
axes[0].set_title('Grafik 2: Tekanan Kepala Sumur vs Ukuran Choke')

sns.scatterplot(x='BORE_OIL_VOL', y='AVG_CHOKE_SIZE_P', data=df_train, alpha=0.5, color='teal', ax=axes[1])
axes[1].set_title('Grafik 3: Volume Minyak vs Ukuran Choke')
plt.show()

# Grafik 4: Scatter Plot Lanjutan (Relationship)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='AVG_DOWNHOLE_PRESSURE', y='AVG_CHOKE_SIZE_P', data=df_train, alpha=0.5, color='purple')
plt.title('Grafik 4: Tekanan Dasar Sumur vs Ukuran Choke')
plt.show()

# Grafik 5 & 6: Boxplot (Comparison)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(y=df_train['AVG_CHOKE_SIZE_P'], color='skyblue', ax=axes[0])
axes[0].set_title('Grafik 5: Distribusi Ukuran Choke')

pressure_data = df_train[['AVG_WELLHEAD_PRESSURE', 'AVG_DOWNHOLE_PRESSURE']]
sns.boxplot(data=pressure_data, ax=axes[1])
axes[1].set_title('Grafik 6: Komparasi Distribusi Tekanan')
plt.show()

# Grafik 7: Tren Time-Series
if 'DATEPRD' in df_train.columns:
    plt.figure(figsize=(14, 5))
    df_sample = df_train.sort_values('DATEPRD').head(500)
    sns.lineplot(x='DATEPRD', y='AVG_CHOKE_SIZE_P', data=df_sample, color='navy')
    plt.title('Grafik 7: Tren Historis Ukuran Choke (Time-Series)')
    plt.xticks(rotation=45)
    plt.show()

# Grafik 8: Pie Chart Komposisi Fluida Cair (Hierarchy)
avg_oil = df_train['BORE_OIL_VOL'].mean()
avg_water = df_train['BORE_WAT_VOL'].mean()
plt.figure(figsize=(7, 7))
plt.pie([avg_oil, avg_water], labels=['Rata-rata Vol Minyak', 'Rata-rata Vol Air'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Grafik 8: Hierarki Rata-rata Produksi Fluida Cair')
plt.show()
```
