```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
print("Import Library Selesai!")
```

```python
# Load data hasil cleansing dari Bab 1
df_train = pd.read_csv('../data/processed/volve_train_clean.csv')

# Gunakan format parameter agar proses konversi tanggal secepat kilat
df_train['DATEPRD'] = pd.to_datetime(df_train['DATEPRD'], format='mixed')

print("Data berhasil dimuat! Jumlah baris:", len(df_train))
```

```python
print("=== BAB 7: VISUALISASI DATA (8 GRAFIK) ===")

# Grafik 1 & 2: Scatter Plot (Relationship)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.scatterplot(x='AVG_WELLHEAD_PRESSURE', y='AVG_CHOKE_SIZE_P', data=df_train, alpha=0.5, color='coral', ax=axes[0])
axes[0].set_title('Grafik 1: Tekanan Kepala Sumur vs Ukuran Choke')

sns.scatterplot(x='BORE_OIL_VOL', y='AVG_CHOKE_SIZE_P', data=df_train, alpha=0.5, color='teal', ax=axes[1])
axes[1].set_title('Grafik 2: Volume Minyak vs Ukuran Choke')
plt.show()

# Grafik 3: Scatter Plot Lanjutan (Relationship)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='AVG_DOWNHOLE_PRESSURE', y='AVG_CHOKE_SIZE_P', data=df_train, alpha=0.5, color='purple')
plt.title('Grafik 3: Tekanan Dasar Sumur vs Ukuran Choke')
plt.show()

# Grafik 4 & 5: Boxplot (Comparison)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(y=df_train['AVG_CHOKE_SIZE_P'], color='skyblue', ax=axes[0])
axes[0].set_title('Grafik 4: Distribusi Ukuran Choke')

pressure_data = df_train[['AVG_WELLHEAD_PRESSURE', 'AVG_DOWNHOLE_PRESSURE']]
sns.boxplot(data=pressure_data, ax=axes[1])
axes[1].set_title('Grafik 5: Komparasi Distribusi Tekanan')
plt.show()

# Grafik 6 & 7: Time-Series
fig, axes = plt.subplots(2, 1, figsize=(14, 10))
df_sample = df_train.sort_values('DATEPRD').head(500)
sns.lineplot(x='DATEPRD', y='AVG_CHOKE_SIZE_P', data=df_sample, color='navy', ax=axes[0])
axes[0].set_title('Grafik 6: Tren Historis Ukuran Choke (Time-Series)')

sns.lineplot(x='DATEPRD', y='BORE_OIL_VOL', data=df_sample, color='darkgreen', ax=axes[1])
axes[1].set_title('Grafik 7: Tren Historis Produksi Minyak (Time-Series)')
plt.tight_layout()
plt.show()

# Grafik 8: Pie Chart Komposisi Fluida Cair (Hierarchy)
avg_oil = df_train['BORE_OIL_VOL'].mean()
avg_water = df_train['BORE_WAT_VOL'].mean()
plt.figure(figsize=(7, 7))
plt.pie([avg_oil, avg_water], labels=['Rata-rata Vol Minyak', 'Rata-rata Vol Air'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Grafik 8: Hierarki Rata-rata Produksi Fluida Cair')
plt.show()
```
